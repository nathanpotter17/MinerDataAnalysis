### Import to DataFrame

```python
#read-in, find all

with open('log20210923_230324.txt','r') as file:
    log1 = file.read()

result = re.findall(r'(Share actual difficulty:\s[0-9]+.....)', log1)
times = re.findall(r'(time: \d+:\d+)', log1)

df1 = pd.DataFrame(result,columns=['Share'])
df2 = pd.DataFrame(times,columns=['Time'])
```

```python
# new data frame with split value columns

df1 = df1.reindex(columns = df1.columns.tolist() + ['MH'] + ['GH'] + ['TH'])
df1['MH']=df1['Share'].str.extract(r'(\d+...\sMH)')
df1['GH']=df1['Share'].str.extract(r'(\d+...\sGH)')
df1['TH']=df1['Share'].str.extract(r'(\d+...\sTH)')
df1['MH'] = df1['MH'].replace(np.nan, 0)
df1['GH'] = df1['GH'].replace(np.nan, 0)
df1['TH'] = df1['TH'].replace(np.nan, 0)
df1.drop(['Share'], axis=1, inplace=True)
df1['MH']=df1['MH'].str.extract(r'(\d+.)')
df1['GH']=df1['GH'].str.extract(r'(\d+...)')
df1['MH'] = df1['MH'].replace(np.nan, 0)
df1['GH'] = df1['GH'].replace(np.nan, 0)
df1=df1.apply(pd.to_numeric, errors='ignore')
mh = df1[['MH']].copy()
mh=mh.loc[~(mh==0).all(axis=1)]
gh = df1[['GH']].copy()
gh=gh.loc[~(gh==0).all(axis=1)]
th = df1[['TH']].copy()
th=th.loc[~(th==0).all(axis=1)]
```

```python
# time oriented data - W.I.P.

time = df2[['Time']].copy()
time['Time']=time['Time'].str.extract(r'(\d+:\d+)')
time = time.loc[~(time=='0:00').all(axis=1)]
df1 = df1.join(time)
df1=df1.apply(pd.to_numeric, errors='ignore')
df1
```

```python
# share distributions - If TH graph is empty, no TH shares were recorded.

sns.set_theme(style="whitegrid", palette="plasma")
sns.displot(data=mh, x="MH", kind="hist")
sns.rugplot(data=mh, x="MH", height=0.1)
sns.displot(data=gh, x="GH", kind="hist")
sns.rugplot(data=gh, x="GH", height=0.1)
sns.displot(data=th, x="TH", kind="hist")
sns.rugplot(data=th, x="TH", height=0.1)
```

```python
#share counting - in order of ascending quality

sns.set_theme(style="whitegrid", palette="plasma")
sns.countplot(data=mh)
```
